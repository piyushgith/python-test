```python
df = pd.read_csv('cricket.csv')
df
```

```python
# Rename the columns to match the original dataset
df.head()
```

```python
new_columns = {'Mat': 'Matches', 'Inns': 'Innings', 'NO': 'Not Out', 'Runs': 'Total Runs', 'HS': 'Highest Score',
               'Ave': 'Batting Average', 'BF': 'Balls Faced', 'SR': 'Strike Rate'}
df.rename(columns=new_columns, inplace=True)
df.head()
```

```python
# Split the Span column into Rookie Year and Final Year
df[['Rookie Year', 'Final Year']] = df['Span'].str.split('-', expand=True)

# Convert the years to numeric type
df['Rookie Year'] = pd.to_numeric(df['Rookie Year'])
df['Final Year'] = pd.to_numeric(df['Final Year'])

# Drop the original Span column if you don't need it anymore
df.drop('Span', axis=1, inplace=True)

# Display the first few rows to verify the changes
df.head()
```

```python
# remove * from the columns values

df['4s'] = df['4s'].str.replace('+', '', regex=False)
df['0'] = df['0'].str.replace('-', '0', regex=False)
df['Balls Faced'] = df['Balls Faced'].str.replace('*', '', regex=False)
df['Highest Score'] = df['Highest Score'].str.replace('*', '', regex=False)
df.info()
```
